```python
from pyspark.sql.functions import *
```

```python
source_name = 'dbfs'
source_path = '/Volumes/zara_lakehouse/retail_source/products_sold/products_sold.csv'
source_file = 'products_sold.csv'
target_table = 'zara_lakehouse.retail_bronze.products_sold_bronze'
```

```python
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(source_path)

```

```python
df = df \
    .withColumn('source', lit(source_name)) \
    .withColumn('file', lit(source_file)) \
    .withColumn("ingestion_time", current_timestamp()) \
    .withColumn("_source_columns", lit(",".join(df.columns)))
```

```python
for col_name in df.columns:
    new_name = col_name.replace(" ", "_")
    lower = new_name.lower()
    df = df.withColumnRenamed(col_name, lower)
```

```python
df = df.write \
        .mode("overwrite") \
        .option ("mergeSchema", "true") \
        .saveAsTable(target_table)
```
