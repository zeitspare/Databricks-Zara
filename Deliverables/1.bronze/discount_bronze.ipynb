```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window
```

```python
path = '/Volumes/zara-lakehouse/retail-source/discount/discount.csv'
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(path)

display(df)
```

```python
duplicates_df = df.groupBy("Product ID", "city") \
    .count() \
    .filter("count > 1")
```

```python
display(duplicates_df)
```

```python
df_filtered = df \
             .filter(F.col("Product ID") == 113143)  \
             .filter(F.col("city") == "San Antonio") \
             .select("Product ID", "city", "discount")
    
display(df_filtered)
```
