# OPTION A

## 1 Filtering Woman Section & Winter Season Fields

- its not sales for winter season sales not filtered in date,  but womean section winter season dress sales

```python
%sql
select  
fs.city,
ROUND(SUM(fs.net_revenue),2) AS total_revenue
from zara_lakehouse.retail_silver.fact_sales_silver fs
JOIN zara_lakehouse.retail_silver.dim_product_information_silver pi
ON fs.product_id = pi.product_id
where pi.section = 'WOMAN' AND  pi.season = 'Winter'
group by fs.city
order by total_revenue desc

```

#### 1b. Sales volume with products of Maximum discount per city

```python
%sql
SELECT
fs.product_id,
MAX(fs.discount) AS max_discount
from zara_lakehouse.retail_silver.fact_sales_silver fs
group by fs.product_id
HAVING max_discount > 0.19
```

```python
%sql

with prod_max_discount AS 
(SELECT
fs.product_id,
MAX(fs.discount) AS max_discount
from zara_lakehouse.retail_silver.fact_sales_silver fs
group by fs.product_id
HAVING max_discount > 0.19)
select 
fs.city,
SUM(fs.net_revenue) AS sales_revenue    
from 
prod_max_discount pmd
JOIN 
zara_lakehouse.retail_silver.dim_product_information_silver pi
on pmd.product_id = pi.product_id
JOIN
zara_lakehouse.retail_silver.fact_sales_silver fs
on pmd.product_id = fs.product_id
where pi.section = 'WOMAN' AND  pi.season = 'Winter'
group by fs.city
order by sales_revenue desc


```

# OPTION B

## 1. Filtering Woman Section with  Winter season dates 
- instead of filtering winter fields from product information table. filtering based on prodcut solds winter season dates.(20 Dec till 20 March )

```python
%sql
select  
fs.city,
ROUND(SUM(fs.net_revenue),2) AS total_revenue
from zara_lakehouse.retail_silver.fact_sales_silver fs
JOIN zara_lakehouse.retail_silver.dim_product_information_silver pi
ON fs.product_id = pi.product_id
where pi.section = 'WOMAN' AND  fs.date BETWEEN '2024-12-20' AND '2025-03-20'
group by fs.city
order by total_revenue desc
```

#### 1b. Sales volume with products of Maximum discount per city

```python
%sql
with prod_max_discount AS 
(SELECT
fs.product_id,
MAX(fs.discount) AS max_discount
from zara_lakehouse.retail_silver.fact_sales_silver fs
group by fs.product_id
HAVING max_discount > 0.19)
select 
fs.city,
ROUND(SUM(fs.net_revenue)) AS sales_revenue    
from 
prod_max_discount pmd
JOIN 
zara_lakehouse.retail_silver.dim_product_information_silver pi
on pmd.product_id = pi.product_id
JOIN
zara_lakehouse.retail_silver.fact_sales_silver fs
on pmd.product_id = fs.product_id
where pi.section = 'WOMAN' AND  fs.date BETWEEN '2024-12-20' AND '2025-03-20'
group by fs.city
order by sales_revenue desc
```
