```python
%sql
SELECT
    fs.product_id,
    fs.city,

    pi.section,
    pi.season,
    pi.material,
    pi.origin,

    SUM(fs.pieces_sold) AS total_sold,
    ROUND(SUM(fs.net_revenue),2) AS total_revenue,
    AVG(fs.discount) AS avg_discount

FROM zara_lakehouse.retail_silver.fact_sales_silver fs

JOIN zara_lakehouse.retail_silver.dim_product_information_silver pi
ON fs.product_id = pi.product_id

GROUP BY 
    fs.product_id,
    fs.city,
    pi.section,
    pi.season,
    pi.material,
    pi.origin;
```
