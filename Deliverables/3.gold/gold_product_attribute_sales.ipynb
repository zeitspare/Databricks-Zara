```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
gold_product_attribute_sales_target = "zara_lakehouse.retail_gold.gold_product_attribute_sales"
```

```python
%sql
SELECT
    fs.product_id,
    fs.city,
    fs.date,
    pi.section,
    pi.season,
    pi.material,
    pi.origin,

    SUM(fs.pieces_sold) AS total_pieces_sold,
    ROUND(SUM(fs.net_revenue),2) AS total_net_revenue,
    AVG(fs.discount) AS avg_discount

FROM zara_lakehouse.retail_silver.fact_sales_silver fs

JOIN zara_lakehouse.retail_silver.dim_product_information_silver pi
ON fs.product_id = pi.product_id

GROUP BY 
    fs.product_id,
    fs.city,
    fs.date,
    pi.section,
    pi.season,
    pi.material,
    pi.origin;
```

```python
df = spark.sql("""
SELECT
    fs.product_id,
    fs.city,
    fs.date,
    pi.section,
    pi.season,
    pi.material,
    pi.origin,

    SUM(fs.pieces_sold) AS total_pieces_sold,
    ROUND(SUM(fs.net_revenue),2) AS total_net_revenue,
    AVG(fs.discount) AS avg_discount

FROM zara_lakehouse.retail_silver.fact_sales_silver fs

JOIN zara_lakehouse.retail_silver.dim_product_information_silver pi
ON fs.product_id = pi.product_id

GROUP BY 
    fs.product_id,
    fs.city,
    fs.date,
    pi.section,
    pi.season,
    pi.material,
    pi.origin;     
               """)
```

```python
df = df.write \
        .mode("overwrite") \
        .option ("mergeSchema", "true") \
        .saveAsTable(gold_product_attribute_sales_target)
```
