```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
gold_city_sales_target = "zara_lakehouse.retail_gold.gold_city_sales"
```

```python
%sql
SELECT
    fs.city,
    SUM(fs.pieces_sold) AS total_pieces_sold,
    ROUND(SUM(fs.net_revenue),2) AS total_net_revenue,
    Round(AVG(fs.discount),2) AS avg_discount
FROM zara_lakehouse.retail_silver.fact_sales_silver fs
GROUP BY fs.city;
```

```python
df = spark.sql("""
SELECT
    fs.city,
    SUM(fs.pieces_sold) AS total_pieces_sold,
    ROUND(SUM(fs.net_revenue),2) AS total_net_revenue,
    Round(AVG(fs.discount),2) AS avg_discount
FROM zara_lakehouse.retail_silver.fact_sales_silver fs
GROUP BY fs.city;
               
               """)
```

```python
df = df.write \
        .mode("overwrite") \
        .option ("mergeSchema", "true") \
        .saveAsTable(gold_city_sales_target)
```
